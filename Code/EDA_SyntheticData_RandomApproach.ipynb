### ***Correlation Determination***

```python
corr_values =  df[["H2", "CH4", "C2H2", "C2H4", "C2H6"]].corr()
```

```python
corr_values
```

```python
chart = sns.heatmap(corr_values, annot=True)
chart.set_title("Correlation between different dissolved gases")
plt.show()
```

The above heatmap shows the below observations:
1. There is a negative correltion between the values which shows that when some gases increses the other gases decrease which is not practically possible.

Hence, this data cannot be considered as anywhere close to actual data.
