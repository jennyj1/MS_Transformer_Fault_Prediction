# Generation of Consolidated Open Source Dataset

```python
#Importing necessary libraries
import os
import openpyxl
import pandas as pd
```

```python
def get_pathname(foldername):
    """ Returns the pathname with the passed foldername where dataset is stored"""
    pathname = os.getcwd()
    pathname = pathname.split("\\")
    pathname = pathname[:-1]
    pathname = "\\".join(pathname)
    pathname = pathname + "\\Dataset\\" + foldername
    return pathname
```

```python
def get_dataframe(pathname, excel_workbook_name):
    """ Takes the pathname and excel workbook name as input and returns a dataframe with data"""
    filename = pathname + "\\" + excel_workbook_name
    data = pd.ExcelFile(filename)
    sheetnames = data.sheet_names
    df = data.parse(sheetnames[0], index = False)
    if not df.empty:
        return df
```

```python
def print_count(dfs):
    """Prints the counts of dataset present in each of the input excel sheets"""
    print("The number fo data resent in wach of the excel sheets from each of the source is as follows:")
    for i in dfs:
        print(len(i.index))
    
```

```python
if __name__ == "__main__":
    foldername = "Open_Source_Dataset"
    pathname = get_pathname(foldername)
    excel_workbook_names = os.listdir(pathname)
    dfs = []
    for wb_name in excel_workbook_names:    
        dfs.append(get_dataframe(pathname, wb_name))
    print_count(dfs)
    consolidated_df = pd.concat(dfs, axis=0)
    print("Dataframe after consolidation")
    display(consolidated_df)
    consolidated_df.drop_duplicates(inplace = True)
    consolidated_df.reset_index(drop = True, inplace = True)
    print("Dataframe after removing duplicates")
    display(consolidated_df)
    print("Displaying all the rows with duplicate data")
    display(consolidated_df[consolidated_df[["H2", "CH4", "C2H2", "C2H4", "C2H6"]].duplicated() == True])
    consolidated_df.to_csv("Consolidated_Actual_Fault_Data.csv", index = False)
    
```
