### ***Correlation Determination***

# **Among gases**

```python
corr_values =  df[["H2", "CH4", "C2H2", "C2H4", "C2H6"]].corr()
```

```python
corr_values
```

```python
chart = sns.heatmap(corr_values, annot=True)
chart.set_title("Correlation between different dissolved gases")
plt.show()
```

The above heatmap shows the below observations:
1. H2 is highly correlated with CH4 and C2H6 and least correlated with C2H2.
2. CH4 is highly correlated with H2 annd C2H6 and least correlated with C2H2.
3. C2H2 is least correlated with H2, C2H6 and CH4 but is highly correlated with C2H4.
4. C2H4 is highly corelated with C2H2
5. C2H6 and H2 have highest correlation followed by CH4 and is least correlated with C2H2.
